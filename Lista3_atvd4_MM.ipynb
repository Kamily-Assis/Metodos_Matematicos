```python
def atualizar_populacao(A, B, C):
    #atualização anual usando as equações iniciais da questão
    novo_A = 0.9 * A + 0.05 * C
    novo_B = 0.8 * B + 0.1 * A
    novo_C = 0.95 * C + 0.2 * B
    return novo_A, novo_B, novo_C

#valores iniciais aleatorios para as populações e ano
A, B, C = 8000, 16000, 11000
anos = 100

#população anual
populacoes_A = [A]
populacoes_B = [B]
populacoes_C = [C]

#passagem dos anos
for ano in range(anos):
    A, B, C = atualizar_populacao(A, B, C)
    populacoes_A.append(A)
    populacoes_B.append(B)
    populacoes_C.append(C)

#registro
for n, (An, Bn, Cn) in enumerate(zip(populacoes_A, populacoes_B, populacoes_C), start=1):
    print(f"Ano {n}: A = {An:.2f}, B = {Bn:.2f}, C = {Cn:.2f}")


#resultados esperados: população com valores que coincidam para (i) -> (0.5, 0.25, 1) -> (10k, 5k, 20k)
```
